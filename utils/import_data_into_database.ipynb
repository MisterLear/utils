```python
import pymysql
import csv
import codecs
```

```python
def get_conn():
    conn = pymysql.connect(host='localhost', port=3306, 
                           user='root', passwd='123456', db='citic', charset='utf8')
    return conn
```

```python
def insert(cur, sql, args):
    cur.execute(sql, args)
```

```python
def load_csv_to_mysql(filename):
    with codecs.open(filename=filename, mode='r', encoding='utf-8') as f:
        reader = csv.reader(f)
        head = next(reader)
        print(head)
        conn = get_conn()
        cur = conn.cursor()
        values_num = "%s, " * len(head)
        values_num = values_num[:-2]
#       sql = 'insert into graduates values(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'
        sql = 'insert into graduates values(%s)'%(values_num)
        for item in reader:
            args = tuple(item)
            print(args)
            insert(cur, sql=sql, args=args)

        conn.commit()
        cur.close()
        conn.close()
```

```python
if __name__ == '__main__':
    load_csv_to_mysql('Desktop/graduates.csv')
```
